```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv('Diwali Sales Data.csv', encoding='unicode_escape')
```

```python
df.shape
```

```python
df.sample(5)
```

```python
df.info()
```

```python
#dropping last two columns as they are empty
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
#checking for null values in each column
pd.isnull(df).sum()
```

```python
#dropping null values
df.dropna(inplace=True)
```

```python
df['Amount']=df['Amount'].astype('int')
```

```python
df['Amount'].dtype
```

```python
df.columns
```

```python
#renaming column

df.rename(columns={"Marital_Status":"Married"})
```

```python
df[['Age', 'Orders', 'Amount']].describe()
```

## EDA


```python
ax=sns.countplot(x="Gender", data=df)

#for showing the values on top of the bars


for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gender=df.groupby(["Gender"], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
ax=sns.barplot(x='Gender', y='Amount', data=sales_gender)

```

```python
ax=sns.countplot(data=df, x='Age Group', hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Total amount by Age Group

sales_age= df.groupby(["Age Group"], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Age Group', y='Amount', data=sales_age)
```

## State vs Orders


```python
sales_state= df.groupby(["State"], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize' :(18,5)})
sns.barplot(data=sales_state, x='State', y='Orders')
```

## States vs Sales Amount

```python
stateVsAmount= df.groupby('State', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize' :(18,5)})
sns.barplot(data=stateVsAmount, x='State', y='Amount')
```

```python
ax= sns.countplot(data=df, x='Marital_Status')

for bars in ax.containers: 
    ax.bar_label(bars)
```

## Marital Status vs Gender as per Sales Amount

```python
sales_marital= df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending= False)

sns.set(rc={'figure.figsize': (6,5)})
sns.barplot(data=sales_marital, x='Marital_Status', y='Amount', hue='Gender')
```

From the above graph, we can see that the majority of buyers are unmarried, women who have spent the most amount.

## Occupation

```python
sns.set(rc={'figure.figsize': (20,9)})

sns.countplot(data=df, x='Occupation', order=df['Occupation'].value_counts().index)

```

## Sales Amount vs Occupation

```python
salesvsOccu= df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(data=salesvsOccu, x='Occupation', y='Amount')
```

From the above graph, we can see that top 3 occupations buying the most are from IT Sector, Healthcare and Aviation.

## Products


```python
sns.set(rc={'figure.figsize':(30,6)})
ax = sns.countplot(data = df, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
```

From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders')
```

```python
# top 10 most sold products (same thing as above)

fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```

 ## In Conclusion
 
 Married women age group 26-35 yrs from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category
